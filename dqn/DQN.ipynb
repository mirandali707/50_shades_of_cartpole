# Deep Q Network
Like Q-learning, except with two Big Ideas:
1. experience replay: store all prior experiences in a replay buffer; in each SGD step, randomly sample an experience tuple to update the network weights.
2. fixed Q-targets: maintain a separate target network that uses a different set of weights than the ones being updated (which are used to decide waht actons to take); only update the target network every n iters (e.g. 1000).
