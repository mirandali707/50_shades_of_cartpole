# Pure random search
Here we randomly initialize a weight vector $w$ and roll out a trajectory (trial/episode) with it. If that trajectory produced the highest reward we've seen so far, we keep that $w$ as our best $w$, otherwise we do nothing. Then we randomly sample another $w$ vector and try again, and so on and so forth.

```python
import sys
sys.path.append('..')
# TODO replace with your agent template below, and optionally also your train loop
from agent import init_agent, reset_agent, get_action, update, model_name, save_trained_agent, init_trained_agent

from train import train_loop
from utils.plotting import plot_trial_results

# if training:
all_trial_results = train_loop(init_agent, reset_agent, get_action, update, model_name, save_trained_agent)
mean_curve = all_trial_results.mean(axis=0)
std_curve = all_trial_results.std(axis=0)
plot_trial_results(model_name, mean_curve, std_curve)
```

Please note that this ^ is not really a "learning curve" per se -- it's just showing that we are generating trials from i.i.d. selection of weight vectors $w$ (so it looks like random noise).
